Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.res = 250, repr.plot.height = 6, repr.plot.width = 8)

p = ggplot(penguins, aes(x = body_mass_g))
p = p + geom_histogram(binwidth = 100, fill = 'steelblue3', color = 'steelblue4',  alpha=0.8, na.rm = TRUE) 
p = p + facet_grid(rows = vars(year))
p = p + theme_minimal()
p = p + scale_x_continuous(label = comma, name ="Body Mass (g)")
p = p + scale_y_continuous(name ="Number of Penguins")
p = p + labs(title = "Histograms of Penguin Body Mass for measurements taken between 2007-2009")
p
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
options(repr.plot.res = 500, repr.plot.height = 6, repr.plot.width = 5)

p = ggplot(penguins, aes(x = body_mass_g, y = sex))
p = p + geom_jitter(color = 'black', alpha = 0.5, height = 0.25, na.rm = TRUE)
p = p + facet_grid(rows = vars(species))
p = p + theme_minimal()
p = p + scale_x_continuous(label = comma, name ="Body Mass (g)")
p = p + scale_y_discrete(name ="Sex")

p
```
